### Allocating the values from the Alcohol and Malic acid columns

```python
x=data.iloc[:,[0,1]].values
x
```

##  K-Means Method

```python
from sklearn.cluster import KMeans
```

```python
x.shape
```

```python
wcss=[]
for i in range(1,19):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

### The within cluster sum of square values :

```python
wcss
```

```python
plt.plot(range(1,19),wcss)
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.xticks(np.arange(1,19))
plt.show()
```

```python
kmeans=KMeans(n_clusters=8,init='k-means++',random_state=42)
y_kmeans=kmeans.fit_predict(x)
```

```python
plt.scatter(x[y_kmeans==0,0],x[y_kmeans==0,1],s=100,c='red',label='Cluster 1')
plt.scatter(x[y_kmeans==1,0],x[y_kmeans==1,1],s=100,c='green',label='Cluster 2')
plt.scatter(x[y_kmeans==2,0],x[y_kmeans==2,1],s=100,c='blue',label='Cluster 3')
plt.scatter(x[y_kmeans==3,0],x[y_kmeans==3,1],s=100,c='yellow',label='Cluster 4')
plt.scatter(x[y_kmeans==4,0],x[y_kmeans==4,1],s=100,c='black',label='Cluster 5')
plt.scatter(x[y_kmeans==5,0],x[y_kmeans==5,1],s=100,c='cyan',label='Cluster 6')
plt.scatter(x[y_kmeans==6,0],x[y_kmeans==6,1],s=100,c='pink',label='Cluster 7')
plt.scatter(x[y_kmeans==7,0],x[y_kmeans==7,1],s=100,c='brown',label='Cluster 8')
plt.xlabel('Alcohol')
plt.ylabel('Malic Acid')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import silhouette_score
silhouette_score_avg=silhouette_score(x,y_kmeans)
silhouette_score_avg
```

## Agglomerative Clustering

```python
import scipy.cluster.hierarchy as sch
```

```python
dend=sch.dendrogram(sch.linkage(x,method='ward'))
plt.ylabel('Euclidean Distance')
plt.title('Dendrogram')
plt.show()
```

## DBSCAN Method (Density-based Spatial Clustering of Applications with Noise)

```python
from sklearn.neighbors import NearestNeighbors
```

```python
neighbors=NearestNeighbors(n_neighbors=7)
neighbors_fit=neighbors.fit(x)
```

```python
distances,indices=neighbors_fit.kneighbors(x)
```

```python
distances=np.sort(distances,axis=0)
distances=distances[:,1]
plt.plot(distances)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
model=DBSCAN(eps=0.36,min_samples=9)
labels=model.fit_predict(x)
```

```python
print('Cluster labels:',labels)
```

```python
unique_labels,counts=np.unique(labels,return_counts=True)
print('Cluster sizes:',dict(zip(unique_labels,counts)))
```

```python
silhouette=silhouette_score(x,labels)
print('Silhouette Score is',silhouette)
```

```python
plt.figure(figsize=(8, 6))
unique_labels = set(labels)
colors = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]


for label,color in zip(unique_labels,colors):
    if label==-1:
        color=[0,0,0,1]
        
    class_member_mask=(labels==label)    
    xy = x[class_member_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=tuple(color), markeredgecolor='k', markersize=14)

plt.title("DBSCAN Clustering")
plt.show()
```

### From the 3 models, the highest silhouette score is of the k-means++ method
