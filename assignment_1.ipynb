# PANDAS

```python
#Displaying the given dataset
iris_data
```

```python
#Identifying the no. of classifications given in the dataset by the 'unique' function
iris_data['Classification'].unique()
```

```python
#Table showing the data of only Iris-setosa species
set1=iris_data[iris_data['Classification']=='Iris-setosa']
set1
```

```python
#set2 is a dataframe with only the information related to the species Iris-setosa
set2=iris_data[iris_data['Classification']=='Iris-setosa']
#Dispalying only the 'SL' and 'SW' columns of the dataframe set2, and with appropraite x and y axis labelings and a heading, given below is a line graph
set2[['SL','SW']].plot.line(xlim=(0,49),ylim=(1.5,6.0),xlabel='Index No.',ylabel='Measurement',title='Iris-setosa');
```

```python
#Dispalying only the 'PL' and 'PW' columns of the dataframe set2, and with appropraite x and y axis labelings and a heading, given below is a line graph
set2[['PL','PW']].plot(xlim=(0,49),ylim=(-0.5,2.0),xlabel='Index No.',ylabel='Measurement in units',title='Iris-setosa');
```

```python
#ver1 is a dataframe with only the information related to the species Iris-versicolor
ver1=iris_data[iris_data.Classification=='Iris-versicolor']
```

```python
#Here it is clear that the Sepal length of the Iris-versicolor species ranges from around 4.9 to around 7.0 units 
ver1['SL'].plot(xlim=(50,99),ylim=(4.5,7.0),xlabel='Index No.',ylabel='Measurement in units',title='Iris-versicolor');
```

The above given data is displayed below in the given boxplot.

# MATPLOTLIB

```python
#The boxplot shows that majority of the data of the Sepal length of Iris-versicolor species is spread between 
plt.boxplot(ver1['SL'])
plt.xlabel('SL')
plt.ylabel('Measurement in units')
plt.title('Boxplot of Sepal Length of Iris-versicolor')
plt.show()
```

# SEABORN

```python
#Line graph of all the data regarding the species Iris-versicolor
sns.lineplot(data=ver1);
plt.xlabel('Index No.')
plt.ylabel('Measurement in units')
plt.title('Iris-versicolor')
plt.show()
```

```python
#vir1 is a dataframe with only the information related to the species Iris-virginica
vir1=iris_data[iris_data.Classification=='Iris-virginica']
vir1
```
