```python
data['Age'].plot(figsize=(7,7))
```

```python
data['Age'].skew()
```

Since the skew value is between -0.5 and 0.5, we can conclude that the age distribution is approximately symmetric, i.e., the mean, median and mode are nearly the same values.

From the skew value and the above line chart, we can conclude that the Age column is approximately symmetric. So we can fill the null values with any of the 3 measures. Here we choose the mean to fill.

```python
data['Age'].mean()
```

29.69 can be approximated to 29.70

```python
data['Age']=data['Age'].fillna(29.70)
data
```

5. Checking for outliers

```python
data
```

```python
data['Age'].plot(kind='box')
```

```python
Q1=np.percentile(data['Age'],25,method='midpoint')
Q2=np.percentile(data['Age'],50,method='midpoint')
Q3=np.percentile(data['Age'],75,method='midpoint')

print('Q1 is: ',Q1)
print('Q2 is: ',Q2)
print('Q3 is: ',Q3)
```

```python
IQR=Q3-Q1
print(IQR)
```

```python
low_lim=Q1-1.5*IQR
up_lim=Q3+1.5*IQR

print('Lower value: ',low_lim)
print('Upper value: ',up_lim)
```

```python
outliers=[]

for i in data['Age']:
    if(i<low_lim) or (i>up_lim):
        outliers.append(i)
        
print(outliers)        
```

```python
data['Age']=data['Age'].clip(lower=low_lim,upper=up_lim)
```

```python
outliers=[]

for i in data['Age']:
    if(i<low_lim) or (i>up_lim):
        outliers.append(i)
        
print(outliers)    
```

Hence, no outliers in the Age column.
